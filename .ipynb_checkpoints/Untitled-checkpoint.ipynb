```python
import pandas as pd
import seaborn as sns
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
```

```python
df = pd.read_csv('Car_sales.csv')
```

```python
df
```

```python
df.isna().sum()
```

```python
df = df.drop(['__year_resale_value'], axis =1 )
```

```python
df = df.dropna()
```

```python
df.shape
```

```python
df
```

```python
def corre(df, thre):
    corr_cols = set()
    corr = df.corr()
    for i in range(len(corr.columns)):
        for j in range(i):
            if abs(corr.iloc[i,j]) > thre:
                cols = corr.columns[i]
                corr_cols.add(cols)
                
    return corr_cols

corre(df, 0.8)
```

```python
df.Vehicle_type.value_counts()
```

```python
x = df.drop(['Price_in_thousands', 'Latest_Launch', 'Fuel_capacity',
 'Fuel_efficiency',
 'Horsepower',
 'Length',
 'Power_perf_factor'], axis = 1)
y = df['Price_in_thousands']
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size= 0.3)
```

```python
x_test
```

```python
x_test[['Manufacturer', 'Model', 'Vehicle_type']] = x_test[['Manufacturer', 'Model', 'Vehicle_type']].apply(LabelEncoder().fit_transform)
```

```python
x_train[['Manufacturer', 'Model', 'Vehicle_type']] = x_train[['Manufacturer', 'Model', 'Vehicle_type']].apply(LabelEncoder().fit_transform)
```

```python
regr = LinearRegression()
```

```python
regr.fit(x_train, y_train)
```

```python
pred = regr.predict(x_test)
```

```python
pred
```

```python
regr.score(x_train, y_train)
```

```python
regr.score(x_test, y_test)
```
