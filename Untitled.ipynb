```python

df['Latest_Launch'] = pd.to_datetime(df['Latest_Launch'])

df['Launch_Year'] = (df['Latest_Launch']).dt.year


```

```python
df = df.drop(['__year_resale_value', 'Latest_Launch'], axis =1 )
df = df.dropna()
df
```

```python
df.columns
```

```python
df_dummies = pd.get_dummies(df)
df_dummies
```

```python
def corre(df, thre):
    corr_cols = set()
    corr = df.corr()
    for i in range(len(corr.columns)):
        for j in range(i):
            if abs(corr.iloc[i,j]) > thre:
                cols = corr.columns[i]
                corr_cols.add(cols)
                
    return corr_cols

corre(df_dummies, 0.8)
```

```python
y = df_dummies['Price_in_thousands']
df_x = df_dummies.drop(['Price_in_thousands'], axis =1)
```

```python
from sklearn.preprocessing import StandardScaler

stand =  StandardScaler()

stand_df = stand.fit_transform(df_x)
```

```python
from sklearn.decomposition import PCA
pca = PCA(n_components=0.75)
red = pca.fit_transform(stand_df)
red.shape
```

```python
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(red, y, test_size= 0.3, random_state= 42)
```

```python
from sklearn.linear_model import LinearRegression

regr = LinearRegression()
```

```python
regr.fit(x_train, y_train)
```

```python
pred = regr.predict(x_test)
```

```python
from sklearn import metrics
rmse = metrics.root_mean_squared_error(pred, y_test)
print (rmse)
```
